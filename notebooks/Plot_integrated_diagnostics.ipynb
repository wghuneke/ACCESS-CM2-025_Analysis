```python
# Plotting info
fig_path = '/g/data/x77/wgh581/Figures/Figures_ACCESS_CM2/'
ft_size  = 16-4
plt.rcParams.update({'font.size': ft_size})
month_label = ['J','F','M','A','M','J','J','A','S','O','N','D']
```

```python
CLEX_Red = '#DA1E48'
CLEX_Pink = '#F48580'
CLEX_SkyBlue = '#00BDF2'
CLEX_MidBlue = '#008DD0'
CLEX_CobaldBlue = '#0066B3'
CLEX_MidnightBlue = '#113559'
CLEX_Gray = '#4D4D47'
```

```python
model = np.arange(5)
model_name = ['Obs', 'CM2-025', 'CM2-1', 'OM2-025', 'OM2-1']
model_color = ['k', CLEX_CobaldBlue, CLEX_SkyBlue, CLEX_Red, CLEX_Pink]
```

```python
fig = plt.figure(figsize=(10,7))

plt.subplot(3,2,1)
plt.scatter(model, SST, c=model_color)
#plt.axhline(SST[0], color='k', linewidth=0.5)
plt.title('a) SST', loc='left', fontsize=ft_size-1)
plt.ylabel('($^{\circ}$C)')
plt.tick_params(labelbottom=False)
plt.xlim(-0.25,4.5)
plt.ylim(18, 19)
plt.yticks(np.arange(18, 19.1, 0.25))

plt.subplot(3,2,2)
plt.scatter(model, MHT_SH, c=model_color)
plt.scatter(model, MHT_NH, c=model_color)
plt.axhline(0, color='k', linewidth=0.5)
plt.title('b) Meridional heat transport', loc='left', fontsize=ft_size-1)
plt.text(0, 0.5, 'NH', color='k')
plt.text(0, -1, 'SH', color='k')
plt.ylabel('(PW)')
plt.tick_params(labelbottom=False)
plt.xlim(-0.25,4.5)
plt.ylim(-2, 2)

plt.subplot(3,2,3)
plt.scatter(model, DP, c=model_color)
plt.title('c) Drake Passage', loc='left', fontsize=ft_size-1)
plt.ylabel('(Sv)')
plt.tick_params(labelbottom=False)
plt.xlim(-0.25,4.5)
plt.ylim(125, 225)

plt.subplot(3,2,4)
plt.scatter(model, ITF, c=model_color)
plt.scatter(model, Lombok, c=model_color, marker='+')
plt.scatter(model, Ombai, c=model_color, marker='x')
plt.scatter(model, Timor, c=model_color, marker='*')
plt.scatter(0, ITF[0], c='k', label='Total')
plt.scatter(0, Lombok[0], c='k', marker='+', label='Lombok')
plt.scatter(0, Ombai[0], c='k', marker='x', label='Ombai')
plt.scatter(0, Timor[0], c='k', marker='*', label='Timor')
plt.legend(ncol=2, loc=4)
plt.axhline(0, color='k', linewidth=0.5)
plt.title('d) Indonesian Throughflow', loc='left', fontsize=ft_size-1)
plt.ylabel('(Sv)')
plt.tick_params(labelbottom=False)
plt.xlim(-0.25,4.5)

plt.subplot(3,2,5)
plt.scatter(model, TK, c=model_color)
plt.title('e) Total kinetic energy', loc='left', fontsize=ft_size-1)
plt.ylabel('(PJ)')
plt.xlim(-0.25,4.5)
plt.xticks(ticks=model, labels=model_name)

plt.subplot(3,2,6)
plt.scatter(model, SIE_AA, c=model_color)
plt.scatter(model, SIE_A, c=model_color, marker='v')
plt.scatter(0, SIE_AA[0], c='k', label='Antarctica')
plt.scatter(0, SIE_A[0], c='k', marker='v', label='Arctic')
plt.legend(loc=4)
plt.title('f) Sea ice extent', loc='left', fontsize=ft_size-1)
plt.ylabel('(10$^{6}$ km$^{2}$)')
plt.xlim(-0.25,4.5)
plt.xticks(ticks=model, labels=model_name)


# Finished
save_flag = 1
if save_flag == 1:
    fig_name = 'Fig_integrated_diagnostics.png'
    plt.savefig(fig_path + fig_name, dpi=300, bbox_inches='tight', facecolor='white');
```
